```python
import numpy as np
import sympy
import matplotlib.pyplot as plt
import math as math
global a
global b
```

```python
def daniel(a, cont):
  b = a.copy()
  for i in range(0, len(a)-1):
    for j in range(0, len(a)-1):
      if i == 0:
        if j == 0:
          if a[i][j+1]==1:cont+=1 
          if a[i+1][j+1]==1:cont+=1
          if a[i+1][j]==1:cont+=1

        elif j == len(a):
          if a[i][j-1]==1:cont+=1 
          if a[i+1][j-1]==1:cont+=1 
          if a[i+1][j]==1:cont+=1 

      elif i == len(a):
        if j == 0:
          if a[i-1][j]==1:cont+=1
          if a[i-1][j+1]==1:cont+=1
          if a[i][j+1]==1:cont+=1 

        elif j == len(a):
          if a[i-1][j-1]==1:cont+=1 
          if a[i-1][j]==1:cont+=1
          if a[i][j-1]==1:cont+=1 

      elif i>0 and i<len(a) and j > 0 and j < len(a):
        if a[i-1][j-1]==1:cont+=1 
        if a[i-1][j]==1:cont+=1
        if a[i-1][j+1]==1:cont+=1

        if a[i][j-1]==1:cont+=1 
        
        if a[i][j+1]==1:cont+=1 

        if a[i+1][j-1]==1:cont+=1 
        if a[i+1][j]==1:cont+=1 
        if a[i+1][j+1]==1:cont+=1
      
      if a[i][j]==0 and cont == 3:b[i][j] = 1
      elif a[i][j]==1 and cont < 2:b[i][j] = 0
      elif a[i][j]==1 and cont > 3:b[i][j] = 0
      elif a[i][j]==1 and (cont == 2 or cont == 3): b[i][j] = 1
      cont = 0
  a = b.copy()
  return a
```

```python
TAM = 100
arg = math.floor((TAM/2) - 3)
```

```python
def gameoflife(a):
  gen = int(input())
  pop = 0
  cont = 0
  print("Celulas na geração 0:")
  plt.figure(figsize=(7, 7))
  plt.axis("off")
  im = plt.imshow(a, aspect = 'auto')
  plt.show()
  print()
  for C in range(1, gen+1):
    print("Geração:", C)
    a = daniel(a, cont)
    pop = 0
    for i in a:
      for j in i:
        if j == 1:
          pop +=1
    if pop == 0:
      break
    print("População atual:", pop)
    plt.figure(figsize=(7, 7))
    plt.axis("off")
    im = plt.imshow(a, aspect = 'auto')
    plt.show()
    print('______________________________________________________')
  print()
  print("Celulas depois da geração:")
  print()
  plt.figure(figsize=(7, 7))
  plt.axis("off")
  im = plt.imshow(a, aspect = 'auto')
  plt.show()
```

```python
#Bote Vida parada
a = np.zeros((TAM,TAM), dtype=np.uint8)
a[0+arg][0+arg] = a[1+arg][0+arg] = 1
a[0+arg][1+arg] = a[2+arg][1+arg] = 1
a[1+arg][2+arg] = 1
bote = a.copy()
plt.figure(figsize=(7, 7))
plt.axis("off")
im = plt.imshow(bote, aspect = 'auto')
plt.show()
```

```python
#SAPO Oscilador
a = np.zeros((TAM,TAM), dtype=np.uint8)
a[0+arg][1+arg] = a[0+arg][2+arg] = a[0+arg][3+arg] = 1
a[1+arg][0+arg] = a[1+arg][1+arg] = a[1+arg][2+arg] = 1
sapo = a.copy()
plt.figure(figsize=(7, 7))
plt.axis("off")
im = plt.imshow(sapo, aspect = 'auto')
plt.show()
```

```python
#LWSS Oscilador
a = np.zeros((TAM,TAM), dtype=np.uint8)
a[2+arg][0+arg] = a[3+arg][0+arg] = a[4+arg][0+arg] = a[1+arg][1+arg] = a[4+arg][1+arg] = a[4+arg][2+arg] = a[4+arg][3+arg] = a[3+arg][4+arg] = a[1+arg][4+arg] = 1
LWSS = a.copy()
plt.figure(figsize=(7, 7))
plt.axis("off")
im = plt.imshow(LWSS, aspect = 'auto')
plt.show()
```

```python
#diehard  A população zera após 130 gerações 
#Matusalem
a = np.zeros((TAM,TAM), dtype=np.uint8)
a[1+arg][0+arg] = 1
a[1+arg][1+arg] = a[2+arg][1+arg] = 1
a[2+arg][5+arg] = 1
a[0+arg][6+arg] = a[2+arg][6+arg] = 1
a[2+arg][7+arg] = 1
matusalem = a.copy()
plt.figure(figsize=(7, 7))
plt.axis("off")
im = plt.imshow(matusalem, aspect = 'auto')
plt.show()
```
